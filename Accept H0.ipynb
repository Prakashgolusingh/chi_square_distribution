```python
'''
A die was thrown 600 times and the following frequencies were observed

Face:       1   2   3       4       5       6
Frequency: 97   99  97      105     101     101

Test the hypothesis that the die is unbiased.

'''
import numpy as np

# Define onserved frequencies ( from rolling the die 600 times )
observed_frequencies = np.array ( [ 97, 99, 97, 105, 101, 101 ] )
print(f'Observed_frequencies: {observed_frequencies}')

# Define expected frequencies ( uniform distribution for a fair die )
# Consider Null hypothesis  H0 : die is unbiased
# Alternative hypothesis   H1 : die is biased

# Since we have assumed that die is unbiased, for each face we will take frequency as total no. of thrown divided by no. of face.
avg = 600/6

expected_frequencies = np.array( [ 600/6, 600/6, 600/6, 600/6, 600/6, 600/6 ])
print(f'Expected_frequencies: {expected_frequencies}')

import scipy.stats as stats

# calculate the chi-square statistic
chi2, p_value = stats.chisquare( observed_frequencies, expected_frequencies )
print(f'Chi square value = {chi2}')

# Degrees of freedom
df = len( observed_frequencies) -1

# Genrally we take significance level(alpha) as 0.05
alpha = 0.05

# Calculate the critical value
critical_value = stats.chi2.ppf ( 1 - alpha, df)
print(f'Significance level = {alpha}')
print(f'Degree of Freedom = {df}')
print(f'Critical value = {critical_value}')


# Determin whether to accept or reject the null hypothesis
if chi2 < critical_value:
    result = "Accept H0: observed frequencies match the expected uniform distrubution"
else:
    result = "Reject H0: observed frequencies do not match the expected uniform distribution"

#Plot the chi- square distribution
import matplotlib.pyplot as plt

x = np.linspace(0,20,1000)
y = stats.chi.pdf(x, df)
plt.plot(x, y, label = f'Chi-square Distribution (df={df})')

# Mark the test statistics on the graph
plt.axvline(x=chi2, color = 'red', linestyle = '--', label = f'Test statistic (={chi2:.2f})')
plt.axvline(x=critical_value, color = 'green', linestyle = '--', label = f'Test statistic (critical_value={critical_value:.2f})')
plt.legend()
plt.title('Chi-square Test for Goodness of Fit - Rolling a Die')
plt.xlabel("Chi-Square value")
plt.ylabel('Probability Density')
plt.show()

print(end='\n')
print('Chi-Square Value < Critical Value', end='\n')
print(f'{chi2} < {critical_value}', end='\n')
print(result)
```
